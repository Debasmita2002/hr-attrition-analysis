___1. Visualize Attrition :___

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'],
                        bins=[18, 25, 35, 45, 55, 65],
                        labels=['18-25', '26-35', '36-45', '46-55', '55+'])
df['AgeGroup'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='AgeGroup', hue='Attrition', data=df)
plt.title("Attrition by Age Group")
plt.show()
```

___Insight : Employees from 26-35 and 36-45 age group has highest attrition rate.___

```python
df['SalaryGroup'] = pd.cut(df['MonthlyIncome'],
                           bins=[0, 3000, 6000, 10000, 20000],
                           labels=['Low', 'Medium', 'High', 'Very High'])
df['SalaryGroup'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='SalaryGroup', hue='Attrition', data=df)
plt.title("Attrition by Salary Group")
plt.show()
```

___Insight : Employees whose monhly income lies in 0-3000 have highest attrition rate.___

```python
plt.figure(figsize=(12,6))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title("Attrition by Job Role")
plt.xticks(rotation=45)
plt.show()
```

```python
attrition_by_role = pd.crosstab(df['JobRole'], df['Attrition'], normalize='index') * 100

attrition_by_role.sort_values(by='Yes', ascending=False)
```

___Insight : List of top 10 job role that has higher risk of attrition is mentioned above.___

# Final Business Insights

## Key Findings:

1. Employees aged 2635 have the highest attrition rate, indicating early-career turnover risk.

2. Employees in the low salary group show significantly higher attrition, suggesting compensation is a major factor.

3. Sales Executive role has the highest attrition, indicating potential workload or job satisfaction issues.

4. Employees working overtime are more likely to leave, showing work-life balance impact.

5. Employees with low job satisfaction and poor work-life balance have higher attrition probability.

6. Employees who left had lower average salary compared to those who stayed.

7. Employees with fewer years at the company have higher attrition, indicating early disengagement.



## Business Recommendations:

 Improve compensation for low salary employees  
 Reduce excessive overtime  
 Improve work-life balance policies  
 Focus retention efforts on Sales department  
 Improve onboarding and engagement for new employees
